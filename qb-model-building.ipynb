```python
import pandas as pd
import numpy as np

# Load datasets
passing_basic = pd.read_csv('data/The Edge - Passing - Basic Passing.csv')
passing_rushing = pd.read_csv('data/The Edge - Passing - Rushing.csv')
passing_air_yards = pd.read_csv('data/The Edge - Passing - Air Yards.csv')
passing_play_calling = pd.read_csv('data/The Edge - Passing - Play Calling.csv')
passing_pressure = pd.read_csv('data/The Edge - Passing - Pressure.csv')
# receiving_basic = pd.read_csv('data/The Edge - Receiving - Basic Receiving.csv')
# receiving_advanced = pd.read_csv('data/The Edge - Receiving - Advanced Receiving.csv')
# rushing_basic = pd.read_csv('data/The Edge - Rushing - Basic Rushing.csv')
# rushing_advanced_receiving = pd.read_csv('data/The Edge - Rushing - Advanced Receiving.csv')
# rushing_receiving = pd.read_csv('data/The Edge - Rushing - Receiving.csv')
# rushing_advanced = pd.read_csv('data/The Edge - Rushing - Advanced Rushing.csv')
# team_defense_basic = pd.read_csv('data/The Edge - Team Defense - Basic Defense.csv')
# team_defense_qb = pd.read_csv('data/The Edge - Team Defense - vs. QB.csv')
# team_defense_wr = pd.read_csv('data/The Edge - Team Defense - vs. WR.csv')
# team_defense_rb = pd.read_csv('data/The Edge - Team Defense - vs. RB.csv')
# team_defense_te = pd.read_csv('data/The Edge - Team Defense - vs. TE.csv')

dataframes = {
    'passing_basic': passing_basic,
    'passing_rushing': passing_rushing,
    'passing_air_yards': passing_air_yards,
    'passing_pressure': passing_pressure,
    'passing_play_calling': passing_play_calling,
}

# Merge all dataframes on 'PLAYER' column
data = dataframes['passing_basic']
for key, df in dataframes.items():
    if key != 'passing_basic':
        data = pd.merge(data, df, on=['PLAYER', 'TEAM'], how='inner', suffixes=('', f'_{key}'))


# Handle missing values
data[['TEAM']] = data[['TEAM']].fillna('N/A', inplace=True, )
# data.fillna(0, inplace=True)
# data['PLAYER'] = data['PLAYER'].astype(str)
# data['TEAM'] = data['TEAM'].astype(str)
# print(data.dtypes)
# 
# 
# # Ensure unique player names by combining with the team to avoid duplicates
# data['PLAYER_TEAM'] = data['PLAYER'] + '_' + data['TEAM']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load datasets (assume the data is already loaded and preprocessed as per previous examples)
# Example: Assuming 'data' is already a merged DataFrame with relevant features and target variables

# Define features and targets
features = data.drop(columns=['CMP', 'ATT', 'YDS', 'RUSH YDS', 'TD', 'INT', 'PLAYER', 'PLAYER_TEAM'])
targets = data[['CMP', 'ATT', 'YDS', 'RUSH YDS', 'TD', 'INT']]

# Encode categorical variables
features = pd.get_dummies(features, columns=['TEAM'])

# Split the data
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, random_state=42)
```

```python
# Initialize the base model
base_model = LinearRegression()

# Wrap the base model in a MultiOutputRegressor
multi_output_model = MultiOutputRegressor(base_model)

# Train the model
multi_output_model.fit(X_train, y_train)

# Evaluate the model
y_pred = multi_output_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred, multioutput='raw_values')
print(f'Mean Squared Error for each target: {mse}')

# Aggregate mean squared error
overall_mse = mean_squared_error(y_test, y_pred)
print(f'Overall Mean Squared Error: {overall_mse}')
```

```python
def predict_player_stats(player, opponent, home):
    player_data = data[data['PLAYER'] == player]
    
    if player_data.empty:
        print("Player not found in the dataset")
        return
    
    # Prepare input features
    player_data['HOME'] = home
    player_data['OPPONENT'] = opponent
    X_input = player_data.drop(columns=['CMP', 'ATT', 'YDS', 'RUSH YDS', 'TD', 'INT', 'PLAYER', 'PLAYER_TEAM'])
    
    # Encode categorical variables
    X_input = pd.get_dummies(X_input, columns=['TEAM', 'OPPONENT'])
    
    # Ensure the input has the same columns as the training data
    X_input = X_input.reindex(columns=X_train.columns, fill_value=0)
    
    # Predict using the trained multi-output model
    prediction = multi_output_model.predict(X_input)
    
    # Map predictions to target variable names
    target_columns = ['CMP', 'ATT', 'YDS', 'RUSH YDS', 'TD', 'INT']
    predictions = {target: prediction[0][i] for i, target in enumerate(target_columns)}
    
    return predictions

# Example prediction
player = 'Josh Allen'
opponent = 'PIT'
home = True

predicted_stats = predict_player_stats(player, opponent, home)
print(predicted_stats)
```
