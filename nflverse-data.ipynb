```python
import pandas as pd


players = pd.read_parquet('data/players.parquet')

years_to_process = ['2020', '2021', '2022', '2023']
plays = None
for year in years_to_process:
    path = f'data/play_by_play_{year}.parquet'
    tmp_plays = pd.read_parquet(path)
    plays = pd.concat([plays, tmp_plays])
    
```

```python
active_players = players.loc[players['status'] == 'ACT']
```

```python
plays
```

```python
# trim down columns
fantasy_plays = plays[['game_id', 'passer', 'passer_id', 'rusher', 'rusher_id', 'receiver', 'receiver_id', 'pass', 'rush', 'yards_gained', 'fumble', 'touchdown', 'pass_touchdown', 'rush_touchdown', 'rush_attempt', 'pass_attempt', 'yards_after_catch']]
```

```python
# filter down to records we care about

filtered_plays = fantasy_plays.dropna(subset=['passer', 'rusher', 'receiver'], how='all')
filtered_plays
```

```python
# aggregate to a game level / player level
game_agg = filtered_plays.groupby(['game_id', 'passer', 'passer_id', 'rusher', 'rusher_id', 'receiver', 'receiver_id']) \
                         .sum()
```
